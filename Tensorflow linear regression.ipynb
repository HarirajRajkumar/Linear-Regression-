```python
import tensorflow as tf
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('auto-mpg.csv')
```

```python
data.head()
```

```python
data['mpg'][0]
```

```python
from sklearn import preprocessing
```

```python
lb=preprocessing.minmax_scale(data['displacement'])
lb2=preprocessing.minmax_scale(data['weight'])
```

```python
# var_1=data['displacement']
var_1=lb
var_1=var_1.tolist()
```

```python
# var_2=data['weight']
var_2=lb2
var_2=var_2.tolist()
```

```python
target=data['mpg']
target=target.values.tolist()
```

### Linear regression

```python
sess=tf.Session()
```

### Y= W1*x1 + W2*x2 + b

```python
W1=tf.Variable([3],dtype=np.float32)
W2=tf.Variable([3],dtype=np.float32)
b=tf.Variable([-3],dtype=np.float32)

x=tf.placeholder(dtype=np.float32)
x1=tf.placeholder(dtype=np.float32)
act_y=tf.placeholder(dtype=np.float32)
```

```python
linear_model = W1*x + W2*x1 + b
```

```python
squared_error=tf.square(linear_model-act_y)
```

```python
loss=tf.reduce_sum(squared_error)
```

```python
optimizer_train = tf.train.GradientDescentOptimizer(.0001)
```

```python
train=optimizer_train.minimize(loss)
```

```python
init=tf.global_variables_initializer()
```

```python
sess.run(init)
```

```python
sess.run(train,{x:var_1,x1:var_2,act_y:target})
```

```python
sess.run(loss,{x:var_1,x1:var_2,act_y:target})
```

```python
sess.run([W1,W2,b])
```

```python
for i in range(0,1000):
    sess.run(train,{x:var_1,x1:var_2,act_y:target})
    print ('loss >>> ',sess.run(loss,{x:var_1,x1:var_2,act_y:target}))
```
